```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
snp=pd.read_csv('../data/sampled_data.csv')
```

```python
snp.head()
```

```python
Allele_1 = list(snp.Allele1)
Allele_2 = list(snp.Allele2)
```

```python
def most_frequent(Allele):
    Most_Frequent=[]
    for j in range(len(Allele[0])):
        List=[]
        
        for i in range(len(Allele)):
            List.append(Allele[i][j])
            
        Most_Frequent.append(max(set(List), key=List.count))
    return Most_Frequent
```

```python
Most_Frequent = most_frequent(Allele_1)
```

```python
# It takes about 8-10 minutes
def combination(x, Most_Frequent):
    List=[]
    for i in range (len(x.Allele1)):
        if x.Allele1[i]!=x.Allele2[i]:
            List.append(1)
        elif x.Allele1[i]==Most_Frequent[i]:
            List.append(0)
        else:
            List.append(2)
    return List
```

```python
combinated = snp.apply(lambda x: combination(x, Most_Frequent), axis=1)
```

```python
combinated = combinated.rename('combinated')
```

```python
expanded = combinated.apply(pd.Series)
```

```python
expanded = expanded.rename(columns = lambda x : 'r' + str(x))
```

```python
snp=snp.join(expanded)
```

```python
snp.to_csv('Processed_data.csv')
```
